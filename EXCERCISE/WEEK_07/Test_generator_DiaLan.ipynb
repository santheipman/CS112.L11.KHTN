```python
import numpy as np
import itertools
```

```python
def is_k_AND_equal_zero(k_elements):
    
    k_AND = (1<<12)-1
    for element in k_elements:
        k_AND &= element
    
    return (k_AND == 0)
```

```python
def brute_force(arr, k):
    """
    Using brute force to get the output
    """
    idx = np.arange(0, len(arr), 1)
    combinations = itertools.combinations(idx, k)
    
    for com in combinations:
        if is_k_AND_equal_zero(arr[list(com)]):
            explain = f'{arr[list(com)]}'
            return 'YES', explain
            break
    else:
        return 'NO', None
```

```python
def generate_one_test(n, k, low_ai, high_ai):
    
    arr = np.random.randint(low=low_ai, high=high_ai+1, size=n)
    return arr
```

```python
def generate_multiple_tests(low_n, high_n, num_tests, low_ai=0, high_ai=4095):
    ns = np.random.randint(low=low_n, high=high_n, size=num_tests)
    for i in range(len(ns)):
        k = np.random.randint(low=1, high=ns[i]+1)
        arr = generate_one_test(ns[i], k, low_ai, high_ai)
        output, explain = brute_force(arr, k)
        print(f'Input:\n{ns[i]} {k}\n{arr}\nOutput:\n{output}')
        if explain:
            print('Explain:\nk elements are:', explain)
        print()
```

```python
generate_multiple_tests(low_n=1, high_n=20, num_tests=3)
```

## Generate test from distributions

### 10 tests where $0 < n \leq 100$ 

```python
generate_multiple_tests(low_n=1, high_n=100, num_tests=10)
```

### 10 tests where $100 < n \leq 10000$ 

```python
generate_multiple_tests(low_n=100, high_n=10000, num_tests=10)
```

### 10 tests where $10000 < n \leq 20000$ 

```python
generate_multiple_tests(low_n=10000, high_n=20000, num_tests=10)
```
