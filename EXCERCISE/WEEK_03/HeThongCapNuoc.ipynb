**Thực hiện yêu cầu (1): tìm số cạnh ít nhất cần thêm vào đồ thị để đồ thị liên thông**

Để nối $n$ vùng liên thông với nhau ta cần ít nhất $n - 1$ cạnh, như vậy để tính số cạnh ít nhất cần thêm vào đồ thị $G$ thì ta chỉ cần **đếm số TPLT của đồ thị**.

**Thực hiện yêu cầu (2): đếm số cách thêm các cạnh thỏa yêu cầu (1)**

Nếu gộp tất cả các đỉnh có trong một TPLT lại thành một đỉnh duy nhất, nói cách khác ta coi mỗi TPLT là một đỉnh, thì bài toán trở thành tính số cây khung của đồ thị và có thể giải bằng việc áp dụng [công thức Cayley](https://en.wikipedia.org/wiki/Cayley%27s_formula).

Tuy nhiên đối với bài toán của chúng ta, ta thấy rằng khi nối $\mathbf{(I)}$ với $\mathbf{(II)}$ thì ta cần chọn một đỉnh thuộc $\mathbf{(I)}$ và một đỉnh thuộc $\mathbf{(II)}$. Nếu $\mathbf{(I)}$ có $I_1$ đỉnh và $\mathbf{(II)}$ có $I_2$ đỉnh thì số cách chọn sẽ bằng $I_1$ $\times$ $I_2$. Việc áp dụng công thức Cayley gốc sẽ không cho ra kết quả đúng. Thay vào đó ta sẽ sử dụng [công thức được rút ra từ công thức Cayley](https://laptrinhx.com/srm-733-cayley-s-formula-derived-2556835044/) (sẽ trình bày cụ thể hơn ở phần sau). Công thức này cần đầu vào là **số TPLT của đồ thị và số đỉnh của mỗi TPLT đó**.

***Tóm lại: Để giải quyết cả 2 yêu cầu của bài toán, ta cần tính số TPLT của đồ thị và số đỉnh của mỗi TPLT đó.***

## **3. Thuật toán**

### **3.1. Ý tưởng**

Ta dùng cấu trúc dữ liệu [Disjoint Set Union](https://cp-algorithms.com/data_structures/disjoint_set_union.html) để xác định các TPLT trong đồ thị.

**Cấu trúc dữ liệu Disjoint Set Union (DSU)**

- DSU giúp ta xác định mỗi đỉnh trong đồ thị thuộc TPLT nào và số lượng đỉnh của mỗi TPLT là bao nhiêu.
- Với DSU, mỗi TPLT sẽ có một nút gốc, nút gốc này cũng sẽ đại diện cho chính TPLT đó.
- DSU gồm 2 thao tác chính:
    - Tìm nút gốc (`find_root`): thao tác này giúp tìm nút gốc của một đỉnh, điều đó tương đương với việc xác định xem đỉnh này thuộc TPLT nào.
    - Hợp hai tập đỉnh (`union_sets`): trong quá trình nhập các cạnh của đồ thị, nếu xuất hiện cạnh nối TPLT này với TPLT kia thì ta phải hợp chúng lại để tạo thành một TPLT thông mới. Các đỉnh trong TPLT nhỏ hơn (ít đỉnh hơn) sẽ được cập nhật lại nút gốc.
- Như vậy để xác định các TPLT của đồ thị ta thực hiện như sau:
    - Ban đầu khởi tạo mỗi đỉnh là một TPLT, đồng thời đỉnh đó cũng chính là nút gốc của TPLT đó.
    - Với mỗi cạnh được nhập vào, ví dụ cạnh nối A và B, thì ta dùng `find_root` để kiểm tra xem A và B có cùng nút gốc hay không (có thuộc cùng một thành phần liên thông hay không), nếu có thì tiếp tục nhập cạnh tiếp theo, còn nếu không thì ta tiến hành hợp hai TPLT và cập nhật lại nút gốc bằng thao tác `union_sets`.
    - Sau khi nhập xong tất cả các cạnh, ta duyệt qua tất cả các đỉnh để đếm số TPLT bằng cách đếm số đỉnh là nút gốc.
    
Độ phức tạp của thuật toán là $\mathbf{O(NlogN)}$ với $N$ là số đỉnh của đồ thị.

**Tính số cách thêm các cạnh**

Chúng ta sử dụng công thức rút ra từ công thức Cayley:

$S =$ $n^{k–2}$ $\times \prod_{i=1}^{k} s_i$

trong đó: 
- $S$ là tổng số cách thêm các cạnh
- $n$ là số đỉnh của đồ thị
- $k$ là số TPLT
- $s_i$ là số đỉnh của TPLT thứ $i$

Phần chứng minh công thức có thể xem tại [đây](https://laptrinhx.com/srm-733-cayley-s-formula-derived-2556835044/).

Trong cài đặt ta cần sử dụng thêm mảng `component_size` để lưu số đỉnh của các TPLT.

### **3.2. Mã giả thuật toán**

```
initialize root, component_size
for edge in edges:
    u, v = Two vertices of the edge
    if u's root is different from v's root:
        union Set_Containing_u and Set_Containing_v
        update root
        update component_size
```

## **4. Cài đặt**

````python
import math
import sys
sys.setrecursionlimit(100000)

def initial_root(n):
    """
    root -- lưu nút gốc của mỗi thành phần liên thông
    component_size -- số đỉnh của mỗi thành phần liên thông
    
    Ban đầu mỗi đỉnh được xem là một thành phần liên thông
    nên root của một đỉnh chính là đỉnh đó và component_size
    của mỗi thành phần liên thông đều bằng 1.
    """
    
    global root, component_size
    root = []
    component_size = []
    for i in range(0, n+1):
        root.append(i)
        component_size.append(1)

        
def find_root(u):
    """
    Tìm root của đỉnh u
    """
    
    global root
    if u == root[u]:
        return u
    return find_root(root[u])


def union_sets(u, v):
    """
    Hợp hai thành phần liên thông lại với nhau
    và cập nhật lại root
    """
    
    global root
    u = find_root(u)
    v = find_root(v)
    if u != v:
        if component_size[u] >= component_size[v]:
            component_size[u] += component_size[v]
            root[v] = u
        else:
            component_size[v] += component_size[u]
            root[u] = v

def inp():
    """
    Nhập input: với mỗi cạnh được nhập vào
    ta kiểm tra hai đầu mút có thuộc cùng TPLT
    không, nếu có thì ta gọi hàm union_sets để
    hợp hai TPLT đó lại
    """
    
    global n, m, root
    n, m = [int(x) for x in input().split()]
    initial_root(n)
    for _ in range(m):
        x, y  = [int(z) for z in input().split()]	
        if x>0 and x<=n and y>0 and y<=n:
            union_sets(x, y)

            
def main():
    mod_number = int(1e9+7)
    global root, component_size
    
    inp()
    
    if n == 1:
        return 0, 0

    number_method = 1  # số cách
    number_component = 0  # số TPLT

    for i in range(1, n+1):
        if find_root(i) == i:  # i là nút gốc, mỗi nút gốc tương ứng với một TPLT
            number_component += 1
            number_method = ((number_method % mod_number) * (component_size[i] % mod_number)) % mod_number
                                            # nhân các s_i lại với nhau
    if number_component == 1:
        return 0, 0

    # số cách = (tích các s_i) * n ^ (k - 2)
    number_method = ((number_method % mod_number) * (int(pow(n, number_component - 2, mod_number)) % mod_number)) % mod_number

    return number_component - 1, number_method


if __name__ == '__main__':
    ans = main()
    print("{}\n{}".format(ans[0], ans[1]))
````
